```python
vacrange = len(vacs) #  range of all vacancies 
vacancies  = [vacs[x] for x  in range(vacrange)] # all vacancies
```

```python
case1 = [vacancies[i] for i in  range(vacrange) if  ' data analyst' in vacancies[i]['vactitle'].lower()] #какие позиции нравятся / выбирите позиции аналитик данных
case2 = [vacancies[i] for i in  range(vacrange) if  'аналитик данных'  in vacancies[i]['vactitle'].lower()] #какие позиции нравятся / выбирите позиции аналитик данных
```

```python
intersting_vacancies  = case1 + case2 
```

```python
intersting_vacancies = case1 + case2
len(intersting_vacancies) # Сколько вакансий для аналатика данных? / нравятся
```

```python
intersting_vacancies[0]['updated_at']
```

```python
month = datetime.datetime.now() - datetime.timedelta(days=30)
updated_at = datetime.datetime.strptime(intersting_vacancies[0]['updated_at'], '%Y-%m-%d %H:%M:%S.%f') 
lastdateVacancies = [intersting_vacancies[choose] for choose in range(len(intersting_vacancies)) if updated_at > month] #  только  свежие вакансии  за последний месяц
for  i  in range(len(lastdateVacancies)):
  print(f'{lastdateVacancies[i]["vactitle"]} {lastdateVacancies[i]["updated_at"]}\n \
        {lastdateVacancies[i]["vacdescription"]}')
```

```python
usePythonVacancies = [intersting_vacancies[choose] for choose in range(len(intersting_vacancies)) if  'python'  in vacancies[choose]['vacdescription'].lower() ] #  использование питон в работе
for  i  in range(len(usePythonVacancies)):
  print(f'{usePythonVacancies[i]["vactitle"]} {usePythonVacancies[i]["updated_at"]}\n \
        {usePythonVacancies[i]["vacdescription"]}')
```
